```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
# Merging both the datasets to make it less complicated  
movies = movies.merge(credits, on = 'title')
```

```python
movies
```

```python
# Columns important for a recommendation system from the movies data
# Genre
# id
# keywords
# title
# overview
# popularity - not bein included as it is a numeric value
```

```python
# Finally having our required dataset ready
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']] 
movies
```

```python
# Checking for missing values
movies.isnull().sum()
```

```python
# Since there are only three rows containing null values we are dropping those rows since overview is important
movies.dropna(inplace = True)
```

```python
# Checking for duplicate values
movies.duplicated().sum()
# There are no duplicate rows
```

```python
type(movies['genres'].iloc[0])
```

```python
# Changing the way the column genres is to our requirement
# From this : '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# To this: ['Action','Adventure','Fantasy','SciFi']
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
 
```

```python
movies['genres'].head()
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
# Similarly applying the same function on other columns as well
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
# For theh case of cast, we only need the first three cast members
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break    
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break

    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
# Splitting overview to convert it into a list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies
```

```python
# Representation for proper tage without any mess between names for example between Sam Worthington and Sam Mendes
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
# Concatenation of all the changed columsn into column named tags
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast']+ movies['crew']
```

```python
# Making a new dataframe
new_df = movies[['movie_id','title','tags']]
new_df
```

```python
# Converting the columns tags from a list to a str
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
# Converting all the tags to lowercase
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
# Since there are words in tags present like action and actions we will try to reduce that by considering only the root words because this causes repetitive action by considering them in the 5000 most common words below
%pip install nltk
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

new_df['tags'] = new_df['tags'].apply(stem)
```

```python
# Using vectorization to make tags column in such a way that the viewers's search will provide nearest 5 results out of 5000 rows
# We collect the most common "5000" words used in the tags column excluding the "stop words". We make this operation possible using scikit learn

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
# Checking our feature names
cv.get_feature_names_out()
```

```python
# Now we are trying to find out the distances between each vector to find out the similarity
# Using cosine similarity from scikit learn to perform this 
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
# similarity[0] gives the similarity of the first movie with the other movies 
```

```python
# defining our movie recommendation function
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True, key = lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)

```

```python
recommend('Avatar')
```
